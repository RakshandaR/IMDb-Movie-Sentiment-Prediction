```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# Sentences 

sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
# Define the vocabulary size
voc_size=10000

# One hot representation - following is the meaning for the code below

# for each word in sent --> convert to one_hot representation by one_hot[words,voc_size] - which means 
# convert each word in sent to one_hot representation of size specified in voc_size 
# all of this should be in a list


onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

```python
# Word Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences # using to make the length of all the sentences same
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# Feature representation

dim = 10
```

```python
# Building the model
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
# Predicting the model

model.predict(embedded_docs)
```
